```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('아이폰_x_전처리.csv')
df
```

```python
# 아이폰X 64GB,아이폰X 256GB,아이폰XR 64GB,아이폰XR 128GB,아이폰XR 256GB,아이폰XS 64GB,아이폰XS 256GB,아이폰XS 512GB,아이폰XS MAX 64GB,아이폰XS MAX 256GB,아이폰XS MAX 512GB
```

```python
#default
df['brand'] = '애플'
```

```python
#default
df['seires']='아이폰 X'
```

```python
#default
df['model'] = 'X'
```

```python
df
```

```python
df['storage']=np.nan
```

```python
# 64GB검출
idx = df[df['title'].str.contains('64')].index
for i in idx:
    df.at[i, 'storage'] = '64'
```

```python
idx = df[df['coment'].str.contains('64')].index
for i in idx:
    df.at[i, 'storage'] = '64'
```

```python
# 128GB검출
idx = df[df['title'].str.contains('128')].index
for i in idx:
    df.at[i, 'storage'] = '128'
idx = df[df['coment'].str.contains('128')].index
for i in idx:
    df.at[i, 'storage'] = '128'
```

```python
# 256GB검출
idx = df[df['title'].str.contains('256')].index
for i in idx:
    df.at[i, 'storage'] = '256'
idx = df[df['coment'].str.contains('256')].index
for i in idx:
    df.at[i, 'storage'] = '256'
```

```python
df['data_pid']=np.nan
```

```python
df['time']=np.nan
```

```python
df['month']=np.nan
```

```python
df
```

```python
#모델명 추출 시작
df['title']=df['title'].str.lower()
```

```python
# xr검출
idx = df[df['title'].str.contains('xr')].index
for i in idx:
    df.at[i, 'model'] = 'XR'
```

```python
# xs검출
idx = df[df['title'].str.contains('xs')].index
for i in idx:
    df.at[i, 'model'] = 'XS'
```

```python
# xs max검출
idx = df[df['title'].str.contains('max')].index
for i in idx:
    df.at[i, 'model'] = 'XS MAX'
```

```python
df.rename(columns = {'region':'location'}, inplace=True)
```

```python
df.rename(columns = {'seires':'series'}, inplace=True)
```

```python
df = df[['data_pid','location','price','time','title','month','brand','series','model','storage','coment']]
```

```python
df
```

```python
df.to_csv('아이폰X.csv',index=False, encoding="utf-8-sig")
```
