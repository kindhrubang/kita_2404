```python
# Import the Speech-to-Text client library
from google.cloud import speech
from google.protobuf import wrappers_pb2

# Instantiates a client
client=speech.SpeechClient()

# The name of the audio file to transcribe
gcs_uri="gs://teamproject3/audio-files/01강 글의 순서(1)_(고3-기본).mp3"

def transcribe_speech():
  audio=speech.RecognitionAudio(uri=gcs_uri)

  config=speech.RecognitionConfig(
  encoding=speech.RecognitionConfig.AudioEncoding.MP3,
  sample_rate_hertz=48000,
  language_code="ko-KR",
  model="latest_long",
  audio_channel_count=2,
  enable_word_confidence=True,
  enable_word_time_offsets=True,
  )

  # Detects speech in the audio file
  operation=client.long_running_recognize(config=config, audio=audio)

  print("Waiting for operation to complete...")
  response=operation.result(timeout=90)

  for result in response.results:
    print("Transcript: {}".format(result.alternatives[0].transcript))

transcribe_speech()
```

```python
import os
from google.cloud.speech_v2 import SpeechClient
from google.cloud.speech_v2.types import cloud_speech

MAX_AUDIO_LENGTH_SECS = 8 * 60 * 60

def run_local_recognize():
    # Instantiates a client.
    client = SpeechClient()

    # 로컬 오디오 파일 경로
    local_audio_file = "/path/to/your/local/audio-file.mp3"

    # 오디오 파일을 바이너리로 읽기
    with open(local_audio_file, "rb") as audio_file:
        content = audio_file.read()

    config = cloud_speech.RecognitionConfig(
        auto_decoding_config={},
        features=cloud_speech.RecognitionFeatures(
            enable_word_confidence=True,
            enable_word_time_offsets=True,
        ),
        model="long",
        language_codes=["ko-KR", "en-US"],  # 필요한 언어 코드로 설정
    )

    # 오디오 데이터를 RecognitionAudio 객체에 담기
    audio = cloud_speech.RecognitionAudio(content=content)

    # 요청 생성
    request = cloud_speech.RecognizeRequest(
        recognizer="projects/deductive-nexus-429503-f9/locations/global/recognizers/_",  # 올바른 recognizer ID 입력
        config=config,
        audio=audio,
    )

    # 음성 인식 요청
    response = client.recognize(request=request)

    # 결과 출력
    for result in response.results:
        print("Transcript: {}".format(result.alternatives[0].transcript))
        for word_info in result.alternatives[0].words:
            print(f"Word: {word_info.word}, Confidence: {word_info.confidence}, Start Time: {word_info.start_offset}, End Time: {word_info.end_offset}")

# 함수 호출
run_local_recognize()
```

```python
import os
from google.cloud.speech_v2 import SpeechClient
from google.cloud.speech_v2.types import cloud_speech

MAX_AUDIO_LENGTH_SECS = 8 * 60 * 60

def run_batch_recognize():
    # Instantiates a client.
    client = SpeechClient()

    # The output path of the transcription result.
    gcs_output_folder = "gs://teamproject3/transcripts"

    # 로컬 오디오 파일 경로
    local_audio_file = "path/to/your/local/audio/file.mp3"

    config = cloud_speech.RecognitionConfig(
        auto_decoding_config={},
        features=cloud_speech.RecognitionFeatures(
            enable_word_confidence=True,
            enable_word_time_offsets=True,
        ),
        model="long",
        language_codes=["ko-KR", "en-US"],
    )

    output_config = cloud_speech.RecognitionOutputConfig(
        gcs_output_config=cloud_speech.GcsOutputConfig(uri=gcs_output_folder),
    )

    # 로컬 파일을 읽어 RecognitionAudio 객체 생성
    with open(local_audio_file, "rb") as audio_file:
        content = audio_file.read()
    audio = cloud_speech.RecognitionAudio(content=content)

    request = cloud_speech.BatchRecognizeRequest(
        recognizer="projects/deductive-nexus-429503-f9/locations/global/recognizers/_",
        config=config,
        files=[cloud_speech.BatchRecognizeFileMetadata(uri="")],  # 빈 문자열로 설정
        recognition_output_config=output_config,
        audio=audio,  # 로컬 오디오 데이터 추가
    )

    operation = client.batch_recognize(request=request)

    print("Waiting for operation to complete...")
    response = operation.result(timeout=3 * MAX_AUDIO_LENGTH_SECS)
    print(response)

# 함수 호출
run_batch_recognize()
```
