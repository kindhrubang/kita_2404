Task1_0530. 고객 정보 데이터프레임과 구매 내역 데이터프레임을 결합하여 각 고객이름(name) 별 총 지출을 계산하여 출력하세요.

```python
import pandas as pd
import numpy as np
# 고객 정보 데이터프레임 생성
customers = pd.DataFrame({
    'customer_id': [1, 2, 3, 4],
    'name': ['Alice', 'Bob', 'Charlie', 'David']
})

# 구매 내역 데이터프레임 생성
purchases = pd.DataFrame({
    'purchase_id': [101, 102, 103, 104, 105],
    'customer_id': [1, 2, 1, 3, 2],
    'amount': [250, 150, 200, 300, 350]
})



df = pd.merge(left = customers  , right = purchases, how = "inner", on = "customer_id")
df1 = df.groupby('name')['amount'].sum()
df1
```

Task2_0530. 학생 성적 데이터프레임과 과목 데이터프레임을 결합하여 각 과목별 평균 성적을 계산하여 출력하세요.

```python
import pandas as pd

# 학생 성적 데이터프레임 생성
grades = pd.DataFrame({
    'student_id': [1, 2, 3, 4, 1, 2, 3, 4],
    'subject_id': [101, 101, 102, 102, 103, 103, 101, 103],
    'grade': [85, 90, 78, 88, 92, 95, 80, 85]
})

# 과목 데이터프레임 생성
subjects = pd.DataFrame({
    'subject_id': [101, 102, 103],
    'subject_name': ['Math', 'Science', 'History']
})

df = pd.merge(left = grades  , right = subjects, how = "inner", on = "subject_id")
df1= df.groupby('subject_id')[['grade']].mean()
df1
df2 = pd.merge(left = subjects, right = df1, how="inner", on ="subject_id")
df2
```

Task3_0530. 고객 데이터프레임에서 age가 30 이상이고 spending이 500 이상인 고객을 찾아 출력하세요

```python
import pandas as pd

# 예제 데이터프레임 생성
customers = pd.DataFrame({
    'customer_id': [1, 2, 3, 4, 5],
    'name': ['Alice', 'Bob', 'Charlie', 'David', 'Eve'],
    'age': [25, 35, 30, 40, 28],
    'spending': [300, 600, 500, 700, 400]
})
df = customers[(customers['age']>=30) & (customers['spending']>=500)]
df
```

Task8_0530. 승객의 가족 크기(동반한 형제자매/배우자 수와 부모/자녀 수의 합)가 생존에 영향을 미치는지를 탐색하여 결과를 출력하세요.

```python
df1=df.copy()
df1['동행자 수']= df.apply(lambda x: x['sibsp'] + x['parch'],axis=1)
df2 = df1.sort_values(by='동행자 수' ,ascending=False)
df2[['name','survived','동행자 수']].head(10)

df2.groupby('동행자 수')[['survived']].mean()
```
