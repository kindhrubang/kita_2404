```python
df['storage']=np.nan
```

```python
# 64GB검출
idx = df[df['title'].str.contains('64')].index
for i in idx:
    df.at[i, 'storage'] = '64'
idx = df[df['coment'].str.contains('64')].index
for i in idx:
    df.at[i, 'storage'] = '64'
```

```python
# 128GB검출
idx = df[df['title'].str.contains('128')].index
for i in idx:
    df.at[i, 'storage'] = '128'
idx = df[df['coment'].str.contains('128')].index
for i in idx:
    df.at[i, 'storage'] = '128'
```

```python
# 256GB검출
idx = df[df['title'].str.contains('256')].index
for i in idx:
    df.at[i, 'storage'] = '256'
idx = df[df['coment'].str.contains('256')].index
for i in idx:
    df.at[i, 'storage'] = '256'
```

```python
# 512GB검출
idx = df[df['title'].str.contains('512')].index
for i in idx:
    df.at[i, 'storage'] = '512'
idx = df[df['coment'].str.contains('512')].index
for i in idx:
    df.at[i, 'storage'] = '512'
```

```python
#모델명 추출 시작
df['title']=df['title'].str.lower()
```

```python
# mini 검출
idx = df[df['title'].str.contains('mini')].index
for i in idx:
    df.at[i, 'model'] = 'mini'
idx = df[df['coment'].str.contains('mini')].index
for i in idx:
    df.at[i, 'model'] = 'mini'

idx = df[df['title'].str.contains('미니')].index
for i in idx:
    df.at[i, 'model'] = 'mini'
idx = df[df['coment'].str.contains('미니')].index
for i in idx:
    df.at[i, 'model'] = 'mini'
```

```python
# pro 검출
idx = df[df['title'].str.contains('pro')].index
for i in idx:
    df.at[i, 'model'] = 'pro'
idx = df[df['coment'].str.contains('pro')].index
for i in idx:
    df.at[i, 'model'] = 'pro'

idx = df[df['title'].str.contains('프로')].index
for i in idx:
    df.at[i, 'model'] = 'pro'
idx = df[df['coment'].str.contains('프로')].index
for i in idx:
    df.at[i, 'model'] = 'pro'
```

```python
# pro max검출
idx = df[df['title'].str.contains('맥스')].index
for i in idx:
    df.at[i, 'model'] = 'pro max'
idx = df[df['coment'].str.contains('맥스')].index
for i in idx:
    df.at[i, 'model'] = 'pro max'

idx = df[df['title'].str.contains('max')].index
for i in idx:
    df.at[i, 'model'] = 'pro max'
idx = df[df['coment'].str.contains('max')].index
for i in idx:
    df.at[i, 'model'] = 'pro max'
```
