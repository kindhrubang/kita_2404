```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.support.ui import WebDriverWait
import time
from bs4 import BeautifulSoup
import pandas as pd


# Chrome 옵션 설정
options= Options()
options.add_argument("--start-maximized")
options.add_experimental_option("detach",True) 

# WEBdRIVER 실행
driver = webdriver.Chrome(options=options)

#검색 키워드 입력
search = input("검색어 입력 :" )

# Google 웹사이트 열기
driver.get('https://www.daangn.com/')

save=[]
df=pd.DataFrame(columns=['title','coment','region','price'])
try:
    driver.find_element(By.XPATH, '//*[@id="gnb-root"]/div/div/div/div/span[1]/form/input').send_keys(search+'\n')
    time.sleep(0.45)
    for i in range(1000):
        driver.find_element(By.XPATH, '//*[@id="result"]/div[1]/div[2]/span').click()
        time.sleep(1)
    soup = BeautifulSoup(driver.page_source)
    df=pd.DataFrame(columns=['title','coment','region','price'])
    pack = soup.select('div.article-info')
    i=0
    box=[]
    for p in pack:
        title = p.select_one('span.article-title').text
        content = p.select_one('span.article-content').text
        region = p.select_one('p.article-region-name').text
        tel = p.select_one('p.article-region-name')
        price = tel.next_sibling.next_sibling.text
        box = [title,content,region,price]
        df.loc[i]=box
        i+=1


    
except Exception as e:
    # 예외 발생 시 메시지 출력
    print("An error occurred:",e)
finally:
    # WebDriver 종료
    driver.quit()
    print('종료(수정)')
```

```python
#price 전처리
df['price']=df['price'].str.replace('\n','')
df['price']=df['price'].str.replace(' ','')
df['price']=df['price'].str.replace(',','')
df['price']=df['price'].str.replace('만원','0000')
df['price']=df['price'].str.replace('원','')
#df['price']=df['price'].astype(int) 가격없음, 나눔
```

```python
#region 전처리
df['region']=df['region'].str.replace('\n','')
```

```python
#coment 전처리
df['coment']=df['coment'].str.replace('\n',' ')
```
