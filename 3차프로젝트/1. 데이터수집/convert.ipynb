```python
#모든파일 한번씩 읽기
import os
import re

# 현재 디렉토리의 모든 파일 목록 가져오기
files = os.listdir()

# .txt 파일만 필터링
txt_files = [f for f in files if f.endswith('.txt')]

for file_name in txt_files:
    try:
        # 파일 읽기
        with open(file_name, 'r', encoding='cp949') as file:
            text = file.read()

        text = re.sub(r'<[^>]+>', '', text)
        with open(file_name, 'w', encoding='utf-8') as file:
            file.write(text)
        print(f"{file_name} 파일 내용:")
        print(text[:200])  # 처음 200자만 출력
        print("...")
        print()
        
    except Exception as e:
        print(f"{file_name} 파일을 읽는 중 오류 발생: {str(e)}")
        print()

print("모든 파일 읽기 완료")

```
