Task3_0528. 주어진 DataFrame의 여러 열에 대해 각기 다른 Series를 더하고, 결과를 새로운 DataFrame으로 반환한 후 각 행의 합계를 계산하여 새로운 열에 추가하여 출력하세요.

```python
import pandas as pd

# 데이터프레임 생성
df = pd.DataFrame({
    "A": [1, 2, 3, 4, 5],
    "B": [10, 20, 30, 40, 50],
    "C": [100, 200, 300, 400, 500]
})

# Series 생성
s1 = pd.Series([5, 5, 5, 5, 5])
s2 = pd.Series([10, 10, 10, 10, 10])
s3 = pd.Series([15, 15, 15, 15, 15])
```

```python
df1= df.transpose()
df1
```

```python
ndf = pd.DataFrame({
    "A+s1":df1.loc['A'] + s1,
    "B+s2":df1.loc['B'] + s2,
    "C+s3":df1.loc['C'] + s3
})
ndf.loc[:,"sum"] = 0
for i in range(5):
    for r in range(3):
        ndf.loc[i,"sum"] = ndf.loc[i,"sum"]+ndf.loc[i][r]
ndf
```

```python
#답안 예시
df_new = df.copy()
df_new['A'] = df['A'] + s1
df_new['B'] = df['B'] + s2
df_new['C'] = df['C'] + s3

# 각 행의 합계를 계산하여 새로운 열 'Sum'에 추가
df_new['Sum'] = df_new.sum(axis=1)

#결과출력
print(df_new)
```

Task6_0528. df에서 두 개의 열을 더한 값이 다른 한 개의 열보다 큰 경우에는 1, 작은 경우에는 0으로 값을 정하는 새로운 열을 생성한 후 출력하세요.

```python
import pandas as pd

# 데이터프레임 생성
df = pd.DataFrame({
    "A": [2, 3, 4, 5, 6],
    "B": [1, 2, 3, 4, 5],
    "C": [3, 4, 5, 6, 7],
    "D": [8, 9, 10, 11, 12],
    "E": [10, 20, 30, 40, 50]
})
```

```python
ndf = df.copy()
ndf.loc[:,"A와B를 C와 비교"] = 0
for i in range(5):
    if ndf.loc[i][0] + ndf.loc[i][1]> ndf.loc[i][2]:
        ndf.loc[i,"A와B를 C와 비교"] = 1
ndf
```

```python
#답안 예시
#두 열의 합이 다른 한 열보다 큰 경우 1, 작거나 같은 경우 0으로 설정하는 새로운 칼럼 새성
```

```python
df['C랑 비교'] = (df['A'] + df['B'] > df['C']).astype(int)

print(df)
```
