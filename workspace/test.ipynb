```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.support.ui import WebDriverWait
import time
from bs4 import BeautifulSoup

# Chrome 옵션 설정
options= Options()
options.add_argument("--start-maximized")
options.add_experimental_option("detach",True) 

# WEBdRIVER 실행
driver = webdriver.Chrome(options=options)

#검색 키워드 입력
search = input("검색어 입력 :" )

# Google 웹사이트 열기
driver.get('https://www.daangn.com/')

title_save = []
price_save = []
try:
    driver.find_element(By.XPATH, '//*[@id="gnb-root"]/div/div/div/div/span[1]/form/input').send_keys(search+'\n')
    time.sleep(0.45)
    for i in range(300):
        driver.find_element(By.XPATH, '//*[@id="result"]/div[1]/div[2]/span').click()
        time.sleep(0.20)


    soup = BeautifulSoup(driver.page_source)


    title = soup.select('span.article-title')
    price = soup.select('p.article-price')
    for t in title:
        title_save.append(t.text)
    for p in price:
        price_save.append(p.text)
except Exception as e:
    # 예외 발생 시 메시지 출력
    print("An error occurred:",e)
finally:
    # WebDriver 종료
    # driver.quit()
    print('종료(수정)')
```
