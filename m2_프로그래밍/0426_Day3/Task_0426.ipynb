Task2_0426. 두 개의 리스트에서 공통으로 나타나는 요소들을 찾아 리스트로 반환하는 프로그램을 작성하세요.

예시 입력: list1 = [1, 2, 3, 4, 5], list2 = [4, 5, 6, 7, 8]

예시 출력: [4, 5]

```python
list1 = [1, 2, 3, 4, 5]
list2 = [4, 5, 6, 7, 8]

def a_and_b(l1, l2):
    set_l1 = set(l1)
    set_l2 = set(l2)

    return set_l1 & set_l2

print(a_and_b(list1,list2))
```

Task3_0426. 주어진 숫자 리스트에서 특정 값보다 큰 요소가 하나라도 존재하는지 검사하고, 그 결과를 불 값으로 반환하는 함수를 작성하세요.

- numbers = [1, 2, 3, 10, 20]
- threshold = 15

```python
numbers = [1, 2, 3, 10, 20]
threshold = 15

def upper_test(num_main, sub):
    for num in num_main:
        if num > sub:
            return True

print(upper_test(numbers,threshold))
```

Task6_0426. Calculator 클래스를 작성하고 4칙연산을 수행하는 객체 4개를 작성하여 결과를 출력하세요.

```python
class calculating_machine():
    def add(self, a, b):
        return a+b
    def sub(self, a, b):
        return a-b
    def mul(self, a, b):
        return a*b
    def par(self, a, b):
        return a/b
cal = calculating_machine()

a = int(input("숫자를 입력하세요 : "))
b = int(input("숫자를 입력하세요 : "))
c = (input("원하는 계산 방식을 입력하세요 더하기/빼기/곱하기/나누기 : "))
while True:
    if c == '더하기':
        print(cal.add(a,b))
        break
    elif c == '빼기':
        print(cal.sub(a,b))
        break
    elif c == '곱하기':
        print(cal.mul(a,b))
        break
    elif c =='나누기':
        print(cal.par(a,b))
        break
    else:
        print('다시 시도하세요')
        break
```

Task7_0426. 다음 과제를 수행하세요.
- 사용자로 부터 텍스트를 입력 받는다.
- 문자을 단어 단위로 분리 : split()
- 단어의 빈도수를 저장할 딕셔너리를 생성
- 각 단어의 빈도 수를 계산(for 문 / if else문)
- 결과 출력

[ 예시 ]

문장을 입력하세요: I love apple. I love orange. Apple is tasty

{'i': 2, 'love': 2, 'apple.': 1, 'orange.': 1, 'apple': 1, 'is': 1, 'tasty': 1}

```python
main_sent = input('문장을 입력하세요')
def same_things(sent):
    sent=sent.lower()
    base = sent.split()
    base_set = set(base)
    base_set_to_list = list(base_set)
    count = 0
    output = {}
    for a in base_set_to_list:
        for b in base:
            if a==b.lower():
                count += 1

        output[a.lower()] = count
        count = 0
    return output

same_things(main_sent)
```
