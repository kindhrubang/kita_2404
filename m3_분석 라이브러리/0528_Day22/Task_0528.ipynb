Task1_0528. 연산메소드 이용 및 NaN 값을 0으로 대체 후 사칙연산을 수행한 후 결과를 출력하세요.

```python
# DataFrame과 Series간 연산
frame = pd.DataFrame(np.arange(12.).reshape(4,3),columns=list('bde'))
frame
```

```python
st1 = pd.DataFrame(np.arange(12.).reshape(4,3),columns=list('bde'))
st2 = st1.iloc[1]

st1 = st1.reindex(fill_value = 0)
st2 = st2.reindex(fill_value=0)

ad = st1.add(st2,fill_value=0)
su = st1.sub(st2,fill_value=0)
mu = st1.mul(st2,fill_value=0)
di = st1.div(st2,fill_value=0)


df = pd.DataFrame([ad,su,mu,di],
                     index=['덧셈','뺄셈','곱셈','나눗셈'])
df
```

Task3_0528. 주어진 DataFrame의 여러 열에 대해 각기 다른 Series를 더하고, 결과를 새로운 DataFrame으로 반환한 후 각 행의 합계를 계산하여 새로운 열에 추가하여 출력하세요.

```python
import pandas as pd

# 데이터프레임 생성
df = pd.DataFrame({
    "A": [1, 2, 3, 4, 5],
    "B": [10, 20, 30, 40, 50],
    "C": [100, 200, 300, 400, 500]
})

# Series 생성
s1 = pd.Series([5, 5, 5, 5, 5])
s2 = pd.Series([10, 10, 10, 10, 10])
s3 = pd.Series([15, 15, 15, 15, 15])
```

```python
df1= df.transpose()
df1
```

```python
ndf = pd.DataFrame({
    "A+s1":df1.loc['A'] + s1,
    "B+s2":df1.loc['B'] + s2,
    "C+s3":df1.loc['C'] + s3
})
ndf.loc[:,"sum"] = 0
for i in range(5):
    for r in range(3):
        ndf.loc[i,"sum"] = ndf.loc[i,"sum"]+ndf.loc[i][r]
ndf
```

Task6_0528. df에서 두 개의 열을 더한 값이 다른 한 개의 열보다 큰 경우에는 1, 작은 경우에는 0으로 값을 정하는 새로운 열을 생성한 후 출력하세요.

```python
import pandas as pd

# 데이터프레임 생성
df = pd.DataFrame({
    "A": [2, 3, 4, 5, 6],
    "B": [1, 2, 3, 4, 5],
    "C": [3, 4, 5, 6, 7],
    "D": [8, 9, 10, 11, 12],
    "E": [10, 20, 30, 40, 50]
})
```

```python
df
```

```python
ndf = df.copy()
ndf.loc[:,"A와B를 C와 비교"] = 0
for i in range(5):
    if ndf.loc[i][0] + ndf.loc[i][1]> ndf.loc[i][2]:
        ndf.loc[i,"A와B를 C와 비교"] = 1
ndf
```
