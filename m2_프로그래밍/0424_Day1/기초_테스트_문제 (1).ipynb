Q10. 숫자를 0으로 나누려 할 때 발생하는 예외를 처리하는 코드를 작성하세요.

```python
def log(num1,num2):
  output=0
  output = num1 / num2

  return output

a = int(input("넣을 숫자를 입력하세요"))
b = int(input("나눌 숫자를 입력하세요"))

if b==0 :
    print("0을 제외한 숫자를 입력해주세요")
else :
    print(log(a,b))
```
