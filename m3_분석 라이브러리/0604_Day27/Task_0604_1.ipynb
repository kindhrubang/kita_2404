```python
dates = pd.date_range('2020/01/01','2024/12/31', freq='QS')
dates1 = pd.date_range('2020/01/01','2024/12/31', freq='Q')
dates

df = pd.DataFrame(dates1, index = dates, columns = ['dates1'])
df
```

```python
dates = pd.date_range('2000-01-01','2000-01-10 23:59', freq='12h')
dates
df= pd.DataFrame(dates,index = dates, columns=['dd'] )
df['dd'] = df['dd'].map(lambda x:x['dd'][])
```

```python
import pandas as pd

#시간 범위 생성(0일부터 10일까지 12시간 간격)
timedelta_range = pd.timedelta_range(start = 0, end='100',freq='12H')

#데이터 프레임 생성
df = pd.DataFrame({
    'timedelta':timedelta_range,
    'cumulative_time':[timedelta.total_seconds()/ 3600 for timedelta in timedelta_range]
})
print(df)
```

Task5. 주어진 df에서 월별로 데이터의 합계를 계산하세요.

```python
import pandas as pd

# 데이터프레임 생성
df = pd.DataFrame({
    'date': ['2022-01-01', '2022-01-15', '2022-02-01', '2022-02-15', '2022-03-01'],
    'value': [1, 2, 3, 4, 5]
})
df['date'] = pd.to_datetime(df['date'])
df
```

```python
df['month']=df['date'].dt.to_period('M')
df
```

```python
monthly_sum = df.groupby('month')['value'].sum().reset_index()
print(monthly_sum)
```

Task6. 주어진 데이터프레임에서 주말(토요일과 일요일) 데이터를 필터링하세요.

```python
import pandas as pd

# 데이터프레임 생성
df = pd.DataFrame({
    'date': ['2022-01-01', '2022-01-02', '2022-01-03', '2022-01-08', '2022-01-09'],
    'value': [1, 2, 3, 4, 5]
})
df['date'] = pd.to_datetime(df['date'])
df
```

```python
# 주말 데이터 필터링
df['day_of_week'] = df['date'].dt.dayofweek
```

```python
df
```

```python
weekend_df = df[df['day_of_week']>=5]
weekend_df
```

Task7. 주어진 df에서 분기별 데이터의 평균을 계산하세요.

```python
import pandas as pd

# 데이터프레임 생성
df = pd.DataFrame({
    'date': ['2022-01-01', '2022-02-01', '2022-03-01', '2022-04-01', '2022-05-01'],
    'value': [1, 2, 3, 4, 5]
})
df['date'] = pd.to_datetime(df['date'])

df
```

```python
# 분기별 데이터 평균 계산
df['quarter']= df['date'].dt.to_period('Q')
quarterly_mean = df.groupby('quarter')['value'].mean().reset_index()
print(quarterly_mean)
```

Task11. 주어진 df에서 작업 일자(월요일부터 금요일)만 남기고 필터링하세요.

```python
import pandas as pd

# 데이터프레임 생성
df = pd.DataFrame({
    'date': ['2022-01-01', '2022-01-03', '2022-01-04', '2022-01-05', '2022-01-08', '2022-01-09'],
    'value': [1, 2, 3, 4, 5, 6]
})
df['date'] = pd.to_datetime(df['date'])
df
```

```python
# 작업일자 (월요일부터 금요일) 필터링
df['day_of_week'] = df['date'].dt.dayofweek
workdays_df = df[df['day_of_week']<5]
print(workdays_df)
```
